```python
resnet_data = pd.read_csv("PRED_50_RESNET_D1_BATTERY.csv")
max_n = 50000
plt.plot(resnet_data.iloc[:max_n,1])
plt.plot(resnet_data.iloc[:max_n,2])
plt.legend(["True", "Resnet"])
```

```python
import numpy as np
N = 400
smoothed = np.convolve(resnet_data.iloc[:max_n,2], np.ones((N,))/N, mode='same')
plt.plot(resnet_data.iloc[:max_n,1])
plt.plot(resnet_data.iloc[:max_n,2])
plt.plot(smoothed)
plt.legend(["True", "Resnet", "Smoothed"])
```
