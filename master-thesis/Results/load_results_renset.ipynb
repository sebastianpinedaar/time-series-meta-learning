```python
import pandas as pd
import json
import numpy as np
import sys 
import os
import matplotlib.pyplot as plt
```

```python
files = os.listdir("json_files/")
files = [file  for file in files if "RESNET_0_WOFT_POLLUTION" in file ]
```

```python
df = pd.DataFrame()

for file in files:
    with open("../Results/json_files/"+file) as json_file:
        json_data = json.load(json_file)
        df = pd.concat([df, pd.DataFrame(json_data)])
```

```python
df.groupby(["Experiment id", "Learning rate"]).mean()
```

```python
files = os.listdir("json_files/")
files = [file  for file in files if "RESNET_0_WOFT_HR" in file ]

df = pd.DataFrame()

for file in files:
    with open("../Results/json_files/"+file) as json_file:
        json_data = json.load(json_file)
        df = pd.concat([df, pd.DataFrame(json_data)])
        
df.groupby(["Experiment id", "Learning rate"]).mean()
```

```python
files = os.listdir("json_files/")
files = [file  for file in files if "RESNET_0_WOFT_BATTERY" in file ]

df = pd.DataFrame()

for file in files:
    with open("../Results/json_files/"+file) as json_file:
        json_data = json.load(json_file)
        df = pd.concat([df, pd.DataFrame(json_data)])
        
df.groupby(["Experiment id", "Learning rate"]).agg(["mean", "count"])
```
