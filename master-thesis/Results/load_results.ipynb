```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
noise = [0.1, 0.01, 0.001, 0.0001, 0.00001,0.0]

val_error = []
test_error = []
noise_type = "additive"
for noise_level in noise:
    
    temp_val = []
    temp_test  = []
    data = np.load ("npy_objects/run03_"+noise_type+"_"+str(noise_level*100000)+".npy",  allow_pickle=True)
    dict_data = data.item()
        
    for i in range(5):
        temp_val.append(dict_data[str(i)+"_val"].cpu().data.numpy())
        temp_test.append(dict_data[str(i)+"_test"].cpu().data.numpy())
        
    val_error.append(temp_val)
    test_error.append(temp_test)
    #val_error.append()
#npy_objects/run03_additive_0.0.npy
```

```python
val_error = np.array(val_error)
```

```python
test_error = np.array(test_error)
```
