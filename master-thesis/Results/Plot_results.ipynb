```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xgb_pollution = pd.read_csv("plots/PRED_50_XGBOOST_D0_POLLUTION.csv")
```

```python
resnet_pollution = pd.read_csv("plots/PRED_50_RESNET_D0_POLLUTION.csv")
```

```python

plt.plot(xgb_pollution.iloc[:,2])
plt.plot(resnet_pollution.iloc[:,2])
plt.plot(xgb_pollution.iloc[:,1])


plt.legend([ "XGB", "Resnet", "True"])
```

```python
max_n = 1000
plt.plot(xgb_pollution.iloc[:max_n,1])
plt.plot(xgb_pollution.iloc[:max_n,2])
plt.plot(resnet_pollution.iloc[:max_n,2])
plt.legend(["True", "XGB", "Resnet"])
```

```python
fig = plt.figure(figsize=(40,4))
plt.plot(xgb_pollution.iloc[:max_n,2])
fig = plt.figure(figsize=(40,4))

plt.plot(xgb_pollution.iloc[:max_n,1], c="green")
```

```python
max_n = 700

fig = plt.figure(facecolor=(1, 1, 1))
plt.plot(xgb_pollution.iloc[500:max_n,1])

plt.savefig("figure.pdf")

fig = plt.figure()
plt.grid(linestyle=" ")
plt.plot(xgb_pollution.iloc[500:max_n,2], c="purple")
plt.savefig("figure2.pdf")

fig = plt.figure()
plt.plot(resnet_pollution.iloc[500:max_n,2], c = "brown")
#plt.legend(["True", "XGB", "Resnet"])
plt.savefig("figure3.pdf")
```
