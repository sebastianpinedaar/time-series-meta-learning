```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import load_data as ld
import pickle
```

```python
#creating no-meta-learning datasets

task_size = 50
mode = "meta-learning"
##pollution
train_dataset, validation_dataset, test_dataset = ld.load_data_pollution(window_size = 5, 
                                                                         task_size = task_size, 
                                                                         stride=1, mode=mode)


pickle_out = open("../Data/TRAIN-POLLUTION-W5-T50-ML.pickle", "wb")
pickle.dump(train_dataset, pickle_out)
pickle_out.close()

pickle_out = open("../Data/VAL-POLLUTION-W5-T50-ML.pickle", "wb")
pickle.dump(validation_dataset, pickle_out)
pickle_out.close()

pickle_out = open("../Data/TEST-POLLUTION-W5-T50-ML.pickle", "wb")
pickle.dump(test_dataset, pickle_out)
pickle_out.close()
```
