```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import load_data as ld
import pickle
```

```python
#creating no-meta-learning datasets

task_size = 50
mode = "meta-learning"
M = "ML"

dataset_name = "POLLUTION"
window_size = 5
##pollution
train_dataset, validation_dataset, test_dataset = ld.load_data_pollution(window_size = window_size, 
                                                                         task_size = task_size, 
                                                                         stride=1, mode=mode)

pickle_out = open("../Data/TRAIN-"+dataset_name+"-W"+str(window_size)+"-T"+str(task_size)+"-"+M+".pickle", "wb")
pickle.dump(train_dataset, pickle_out)
pickle_out.close()

pickle_out = open("../Data/VAL-"+dataset_name+"-W"+str(window_size)+"-T"+str(task_size)+"-"+M+".pickle", "wb")
pickle.dump(validation_dataset, pickle_out)
pickle_out.close()

pickle_out = open("../Data/TEST-"+dataset_name+"-W"+str(window_size)+"-T"+str(task_size)+"-"+M+".pickle", "wb")
pickle.dump(test_dataset, pickle_out)
pickle_out.close()
```
